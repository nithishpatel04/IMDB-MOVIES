```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
import os

# Function to combine features
def combine_features(row):
    return f"{row['description']} {row['genre']} {row['directors']} {row['stars']}"

# Load your dataset
def load_data():
    data = pd.read_csv('Movies_IMDB.csv')  # Replace with your file path
    data['combined_features'] = data.apply(combine_features, axis=1)
    return data

# Save data and model to pickle
def save_to_pickle(df):
    df.reset_index(drop=False, inplace=True)  # Ensure 'index' is a column
    pickle.dump(df, open('movie_data.pkl', 'wb'))

    tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 1), min_df=0.0, stop_words='english')
    tfidf_matrix = tf.fit_transform(df['combined_features']).toarray()
    pickle.dump(tf, open('tfidf_vectorizer.pkl', 'wb'))
    pickle.dump(tfidf_matrix, open('tfidf_matrix.pkl', 'wb'))

# Main execution
if __name__ == "__main__":
    df = load_data()
    save_to_pickle(df)
    print("Pickle files created successfully!")
```

```python
import gzip
import pickle

def compress_pickle(input_file, output_file):
    with open(input_file, 'rb') as f_in:
        with gzip.open(output_file, 'wb') as f_out:
            f_out.writelines(f_in)

# Compress your TF-IDF files
compress_pickle('tfidf_vectorizer.pkl', 'tfidf_vectorizer.pkl.gz')
compress_pickle('tfidf_matrix.pkl', 'tfidf_matrix.pkl.gz')
```
